```python
#正規分布によって潜在空間の「だいたいこの辺にボールが集まってくれ」というゆるやかな制約を与える
#lec平均 tf.zerosは要素がすべて0,scale標準偏差
import tensorflow as tf
import tensorflow_probability as tfp
encoded_dim=100
tfd = tfp.distributions
prior = tfd.Independent(
tfd.Normal(loc=tf.zeros(encoded_dim), scale=1),
reinterpreted_batch_ndims=1)
```

```python
#148
seq_length = x_train.shape[2]
#1025
input_dim = x_train.shape[1]
hidden_dim=1500
encoder = tf.keras.Sequential()
#音高軸方向の要素数を 1025 → 1 に変換
encoder.add(tf.keras.layers.Conv2D(hidden_dim, (input_dim,1),#横input_dim,縦1のピクセルのフィルタをhidden_dim枚使って畳み込み
                                   input_shape=(input_dim,seq_length, 1),#横seq_length,縦input_dimの画像と色の数
                                   strides=1, padding="valid",
                                   activation="relu"))

#時間軸方向の要素数を 148 → 4 に変換
encoder.add(tf.keras.layers.Conv2D(hidden_dim, (1,37), strides=( 1,37),
                                   padding="valid", activation="relu"))

#時間軸方向の要素数を 4 → 1 に変換
encoder.add(tf.keras.layers.Conv2D(hidden_dim, (1,4), strides=(1,4),
                                   padding="valid", activation="relu"))

#CNN 用の行列(2 次元配列)を 1 次元配列に変換
encoder.add(tf.keras.layers.Flatten())
encoder.add(tf.keras.layers.Dense(
    tfp.layers.MultivariateNormalTriL.params_size(encoded_dim),
    activation=None))
encoder.add(tfp.layers.MultivariateNormalTriL(
    encoded_dim,
    activity_regularizer=tfp.layers.KLDivergenceRegularizer(
        prior, weight=0.001)))
encoder.summary()
```

```python
decoder = tf.keras.Sequential()
#16 次元の潜在空間のベクトルを 1024 次元ベクトルに変換
decoder.add(tf.keras.layers.Dense(hidden_dim, input_dim=encoded_dim,
                                  activation="relu"))

#CNN 用の行列(2 次元配列)に変換
decoder.add(tf.keras.layers.Reshape((1, 1, hidden_dim)))
#時間軸方向の要素数を 1 → 4 に変換
decoder.add(tf.keras.layers.Conv2DTranspose( hidden_dim, (1,4), strides=(1,4), padding="valid", activation="relu"))
#時間軸方向の要素数を 4 → 148 に変換
decoder.add(tf.keras.layers.Conv2DTranspose( hidden_dim, (1,37), strides=(1,37), padding="valid", activation="relu"))
#周波数軸方向の要素数を 1 → 1025 に変換(要素ごとのベクトルは 1 次元に)
decoder.add(tf.keras.layers.Conv2DTranspose( 1, (input_dim,1), strides=1, padding="valid", activation="relu"))
decoder.summary()
```

```python
from keras.callbacks import ModelCheckpoint
import os

os.makedirs('drive/MyDrive/B4川原/vae_model', exist_ok=True)
vae = tf.keras.Model(encoder.inputs, decoder(encoder.outputs))
vae.load_weights('/content/drive/MyDrive/B4川原/vae_model/model_vol6789.h5')
vae.compile(optimizer="adam", loss="mse", metrics="mse")
model_checkpoint = ModelCheckpoint(
    filepath=os.path.join('drive/MyDrive/B4川原/vae_model', 'model_vol6789.h5'),
    monitor='loss',
    save_best_only=True,
    verbose=1)
# 2500

history = vae.fit(x_train, x_train, epochs=500,batch_size=16,callbacks=[model_checkpoint])
```

```python
import numpy as np
from scipy.io.wavfile import write
import librosa
import os

# ①と②の条件を満たすペアを見つける
def find_pairs(original_S, re_spec):
    high_similarity_pairs = []
    low_similarity_pairs = []

    # ①の条件を満たすペアを探す
    for i in range(len(original_S)):
        for j in range(len(re_spec)):
            cosine_similarity = np.dot(original_S[i].flatten(), re_spec[j].flatten()) / (np.linalg.norm(original_S[i]) * np.linalg.norm(re_spec[j]))
            if cosine_similarity >= 0.9:
                high_similarity_pairs.append((i, j))

    # ②の条件を満たすペアを探す
    pair_count = 0
    for i, j in high_similarity_pairs:
        # if pair_count >= 50:
            # break
        for k in range(len(re_spec)):
            if k != j:
                cosine_similarity = np.dot(re_spec[j].flatten(), re_spec[k].flatten()) / (np.linalg.norm(re_spec[j]) * np.linalg.norm(re_spec[k]))
                if 0.40 <= cosine_similarity < 0.7:
                    low_similarity_pairs.append((j, k))
                    pair_count += 1
                    # if pair_count >= 50:
                        # break

    return low_similarity_pairs





# my_xを生成し、WAVファイルとして保存してアップロードする
def process_and_upload_pairs(pairs, z, decoder, service, folder_id, output_directory, sr):
    processed_pairs = set()
    count = 1


    for v, w in pairs:
        if (v, w) in processed_pairs or (w, v) in processed_pairs:
            continue

        # 3桁の数字としてフォーマット
        v_formatted = f"{v:03d}"
        w_formatted = f"{w:03d}"

        # my_xを生成
        n = 0.5
        my_x = decoder.predict(np.array([z[v]]) * n + np.array([z[w]]) * (1 - n))
        my_x = np.squeeze(my_x)

        # WAVファイルとして保存
        y_inv = librosa.griffinlim(my_x, n_iter=32, win_length=2048, hop_length=int(2048/4))
        output_filename = f'morphed_{v_formatted}_{w_formatted}.wav'
        output_path = os.path.join(output_directory+f"/{count}", output_filename)
        write(output_path, sr, y_inv)

        # WAVファイルとして保存(瀬川)
        # WAVファイルとして保存

        my_v = decoder.predict(np.array([z[v]]))
        my_v = np.squeeze(my_v)


        v_inv = librosa.griffinlim(my_v, n_iter=32, win_length=2048, hop_length=int(2048/4))
        output_filename = f'v{v_formatted}.wav'
        output_path = os.path.join(output_directory+f"/{count}", output_filename)
        write(output_path, sr, v_inv)

        # output_filename = f'v{v_formatted}.wav'
        # output_path = os.path.join(output_directory+f"/{count}", output_filename)
        # write(output_path, sr, z[v])

        # WAVファイルとして保存(瀬川)

        my_w = decoder.predict(np.array([z[w]]))
        my_w = np.squeeze(my_w)


        w_inv = librosa.griffinlim(my_w, n_iter=32, win_length=2048, hop_length=int(2048/4))
        output_filename = f'w{w_formatted}.wav'
        output_path = os.path.join(output_directory+f"/{count}", output_filename)
        write(output_path, sr, w_inv)


        # output_filename = f'w{w_formatted}.wav'
        # output_path = os.path.join(output_directory+f"/{count}", output_filename)
        # write(output_path, sr,z[w])
        count += 1
        print(count)

        # 処理されたペアを記録
        processed_pairs.add((v, w))

```

```python
def find_highest_similarity(my_x, re_spec):
    highest_similarity = -1
    highest_similarity_index = -1

    for i, spec in enumerate(re_spec):
        cosine_similarity = np.dot(my_x.flatten(), spec.flatten()) / (np.linalg.norm(my_x) * np.linalg.norm(spec))
        if cosine_similarity > highest_similarity:
            highest_similarity = cosine_similarity
            highest_similarity_index = i

    return highest_similarity, highest_similarity_index

# find_pairsを呼び出してpairsを取得
pairs = find_pairs(original_S, re_spec)

# 各ペアに対して処理を行う
for v, w in pairs:
    n = 0.5  # 混合係数
    my_x = decoder.predict(np.array([z[v]]) * n + np.array([z[w]]) * (1 - n))
    my_x = np.squeeze(my_x)

    # re_specとの類似度を計算
    highest_similarity, highest_similarity_index = find_highest_similarity(my_x, re_spec[:225])
    print(f"ペア: ({v}, {w}), 最高類似度: {highest_similarity}, re_specのインデックス: {highest_similarity_index}")
```

```python
def calculate_average_highest_similarity(pairs, re_spec, z, decoder):
    total_similarity = 0
    count = 0

    for v, w in pairs:
        n = 0.5  # 混合係数
        my_x = decoder.predict(np.array([z[v]]) * n + np.array([z[w]]) * (1 - n))
        my_x = np.squeeze(my_x)

        # re_specとの類似度を計算
        highest_similarity, _ = find_highest_similarity(my_x, re_spec[:225])
        total_similarity += highest_similarity
        count += 1

    average_similarity = total_similarity / count if count > 0 else 0
    return average_similarity

# 平均最高類似度を計算
average_highest_similarity = calculate_average_highest_similarity(pairs, re_spec, z, decoder)
print(f"平均最高類似度: {average_highest_similarity}")

```

```python
import numpy as np

num_iterations = 100
similarity_sum = 0.0

# 閾値ごとのカウント用のリスト
threshold_counts = [0] * 11  # 0から1までの範囲を10に分割

for _ in range(num_iterations):
    i = np.random.randint(0, len(original_S))
    j = np.random.randint(0, len(original_S))

    while i == j:
        j = np.random.randint(0, len(original_S))

    n = 0.5  # モーフィングの係数

    # スペクトログラムをモーフィング
    my_x = decoder.predict(np.array([z[i]]) * n + np.array([z[j]]) * (1 - n))
    my_x = np.squeeze(my_x)

    a = my_x

    similarities = []
    for k in range(len(original_S)):
        b = original_S[k]
        cosine_similarity = np.dot(a.flatten(), b.flatten()) / (np.linalg.norm(a) * np.linalg.norm(b))
        similarities.append(cosine_similarity)

    similarity_sum += np.mean(similarities)

    # 類似度が閾値ごとにカウント
    for idx, threshold in enumerate(np.arange(0, 1.1, 0.1)):
        next_threshold = round(threshold + 0.1, 1) if idx < 10 else 1.0
        count = np.sum((np.array(similarities) >= threshold) & (np.array(similarities) < next_threshold))
        threshold_counts[idx] += count

average_similarity = similarity_sum / num_iterations
print("Average similarity:", average_similarity)

# 各範囲ごとのカウント結果を表示
for idx, threshold in enumerate(np.arange(0, 1.1, 0.1)):
    next_threshold = round(threshold + 0.1, 1) if idx < 10 else 1.0
    print(f"Number of similarities between {threshold:.1f} and {next_threshold:.1f}: {threshold_counts[idx]}")

# for idx, threshold in enumerate(np.arange(0, 1.1, 0.1)):
    # next_threshold = round(threshold + 0.1, 1) if idx < 10 else 1.0
    # print(f"Number of similarities between {threshold} and {next_threshold}: {threshold_counts[idx]}")

```
